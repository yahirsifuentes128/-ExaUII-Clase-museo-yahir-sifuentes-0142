```python
class Museo:
    def __init__(self, nombre, ciudad, tema):
        self.nombre = nombre
        self.ciudad = ciudad
        self.tema = tema

    def leerdatos(self):
        self.nombre = input("Ingrese el nombre del museo: ")
        self.ciudad = input("Ingrese la ciudad: ")
        self.tema = input("Ingrese el tema principal: ")
        print("---------------------------------------")

    def mostrardatos(self):
        print("Nombre: ", self.nombre)
        print("Ciudad: ", self.ciudad)
        print("Tema:   ", self.tema)

# Creando los 2 objetos (museo1, museo2)
museo1 = Museo("", "", "")
museo2 = Museo("", "", "")

# Procesando datos para museo1
print("DATOS DEL MUSEO 1")
museo1.leerdatos()
museo1.mostrardatos()
print("* _ * _ * _ * _ * _ * _ * _ * _ *")

# Procesando datos para museo2
print("DATOS DEL MUSEO 2")
museo2.leerdatos()
museo2.mostrardatos()
print("* _ * _ * _ * _ * _ * _ * _ * _ *")

print("Software desarrollado por Yahir Sifuentes NL = 58")
```
